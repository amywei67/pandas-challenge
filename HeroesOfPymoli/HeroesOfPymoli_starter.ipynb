### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

### Note

```python
import pandas as pd

file_to_load = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(file_to_load)
print(purchase_data)
```

## Player Count

* Display the total number of players


```python
player_info = purchase_data.loc[:, ["Gender", "SN", "Age"]]
player_info = player_info.drop_duplicates()
num_players = player_info.count()[0]
pd.DataFrame({"Total Players": [num_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = len(purchase_data["Item ID"].unique())
print(item_count)

average_price = purchase_data["Price"].mean()
print (average_price)

total_purchase_value = purchase_data["Price"].sum()
print(total_purchase_value)

purchase_count = purchase_data["Price"].count()

#data frame for result
summary_table = pd.DataFrame({"Total Number of Unique Items": item_count,
                              "Number of Purchases": [purchase_count],
                              "Average Price": [average_price],
                              "Total Purchase Value": [total_purchase_value],})


summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_totals = player_info["Gender"].value_counts()
gender_percents = gender_totals / num_players * 100
gender_info = pd.DataFrame({"Total Count": gender_totals, "Percentage of Players": gender_percents})

gender_info = gender_info.round(2)

gender_info
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_total = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Value")
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_counts = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

Average_Purchase_per_person = gender_total / gender_info["Total Count"]

gender_data = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": gender_average, "Total Value": gender_total, "Average Total Purchase per Person": Average_Purchase_per_person})

gender_data["Total Value"] = gender_data["Total Value"].map("${:,.2f}".format)
gender_data["Avg Total Purchase per Person"] = gender_data["Average Total Purchase per Person"].map("${:,.2f}".format)
gender_data["Average Purchase Price"] = gender_data["Average Purchase Price"].map("${:,.2f}".format)
gender_data ["Purchase Count"] = gender_data["Purchase Count"].map("{:,}".format)
gender_data = gender_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Value", "Avg Total Purchase per Person"]]

gender_data
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

player_info["Age Ranges"] = pd.cut(player_info["Age"], age_bins, labels=group_names)

age_info_totals = player_info["Age Ranges"].value_counts()
age_info_percents = age_info_totals / num_players * 100
age_info = pd.DataFrame({"Total Count": age_info_totals, "Percentage of Players": age_info_percents})

age_info = age_info.round(2)

age_info.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Value")
age_average = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

Avg_total_per_person = age_total / age_info["Total Count"]

age_data = pd.DataFrame({"Purchase Count": age_counts, "Average Purchase Price": age_average, "Total Value": age_total, "Average Purchase Total per Person": Avg_total_per_person})

age_data["Average Purchase Price"] = age_data["Average Purchase Price"].map("${:,.2f}".format)
age_data["Total Value"] = age_data["Total Value"].map("${:,.2f}".format)
age_data ["Purchase Count"] = age_data["Purchase Count"].map("{:,}".format)
age_data["Avg Total Purchase per Person"] = age_data["Average Purchase Total per Person"].map("${:,.2f}".format)
age_data = age_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Value", "Avg Total Purchase per Person"]]

age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
user_total = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Value")
user_average = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")

user_data = pd.DataFrame({"Total Value": user_total, "Average Purchase Price": user_average, "Purchase Count": user_count})

user_sorted = user_data.sort_values("Total Value", ascending=False)


user_sorted["Average Purchase Price"] = user_sorted["Average Purchase Price"].map("${:,.2f}".format)
user_sorted["Total Value"] = user_sorted["Total Value"].map("${:,.2f}".format)
user_sorted = user_sorted.loc[:,["Purchase Count", "Average Purchase Price", "Total Value"]]

user_sorted.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

total_item = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Value")
average_item = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

item_data_pd = pd.DataFrame({"Total Value": total_item, "Item Price": average_item, "Purchase Count": item_count})

item_data_count_sorted = item_data_pd.sort_values("Purchase Count", ascending=False)

item_data_count_sorted["Item Price"] = item_data_count_sorted["Item Price"].map("${:,.2f}".format)
item_data_count_sorted["Purchase Count"] = item_data_count_sorted["Purchase Count"].map("{:,}".format)
item_data_count_sorted["Total Value"] = item_data_count_sorted["Total Value"].map("${:,.2f}".format)
item_popularity = item_data_count_sorted.loc[:,["Purchase Count", "Item Price", "Total Value"]]

item_popularity.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_total = item_data_pd.sort_values("Total Value", ascending=False)

item_total["Item Price"] = item_total["Item Price"].map("${:,.2f}".format)
item_total["Purchase Count"] = item_total["Purchase Count"].map("{:,}".format)
item_total["Total Value"] = item_total["Total Value"].map("${:,.2f}".format)
item_profitable = item_total.loc[:,["Purchase Count", "Item Price", "Total Value"]]

item_profitable.head(5)
```
